####Start build a SVM model on dataset by using important feature 

```python
#Import important libraries that we have reqired while building model
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
#load data
filename = 'forestfires.csv'
names = ['month','RH','DC','DMC','area','size_category']
dataframe = read_csv(filename, names=names, header=0)
```

```python
#Convert data from categrical into numerical 
dataframe[['size_category']]= dataframe[['size_category']].apply(lambda x: pd.factorize(x) [0])
```

```python
dataframe
```

```python
dataframe.info()
```

```python
array = dataframe.values
X = array[:,0:-1]
Y = array[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

####Grid Search CV

```python
#Build a Support Vector Classification model on dataset using kernel= Radial Basis Function(rbf)
from sklearn.svm import SVC
#Find out the best values by using GridSearchCV
#a) gamma(control the influence of new feature mean how much you want the influence of new feature on the data); 
#gamma= higher value[ more influence of that new feature will have on the decision boundary, our model will slowly become from leberal to conservative model= means model have a more controll on the data(more penalty SVM get when it makes missclassification) & it will be less soft margin], 
#gamma= lower value[ less influence of new feature on decision boundary, liberal model= means free model it will tolerate anything (very less penalty SVM get when it makes missclassification)& it will be more soft margin]
#New feature= distance between observations & centre
#Soft margin:- it allows few observation which has made missclassify wrongly
#b)Cost['C'] or Penalty term
from sklearn.model_selection import GridSearchCV
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(X_train,y_train)
```

```python
#We have got best 'C', gamma values with model accuracy; if working with kernel is (rbf)
gsv.best_params_ , gsv.best_score_ 
```

```python
#I want to build a SVC model with C=0.1, gamma=50 & kernel= rbf
from sklearn.metrics import accuracy_score, confusion_matrix
clf = SVC(C= 0.1, gamma = 50, kernel= "rbf")
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```

```python
#kernel= linear
clf = SVC(kernel= "linear") 
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)    
```

```python
#kernel= polynomial
clf = SVC(kernel= "poly") 
clf.fit(X_train , y_train)
y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred) 
```
